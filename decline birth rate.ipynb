```python
import matplotlib.pyplot as plt
import numpy as np

# --- Project Configuration ---

# Set a professional 'dark mode' style similar to the presentation
plt.style.use('dark_background')

# Define the primary colors from the presentation
COLOR_PRIMARY_TEXT = '#CBD5E1'
COLOR_TITLE = '#FFFFFF'
COLOR_ACCENT_1 = '#38BDF8'  # Light Blue
COLOR_ACCENT_2 = '#FACC15'  # Yellow
COLOR_ACCENT_3 = '#F87171'  # Red/Pink
COLOR_ACCENT_4 = '#9CA3AF'  # Grey
COLOR_GRID = '#334155'
COLOR_BG = '#0F172A'
COLOR_BAR_TEXT = '#0F172A'

# Set default font properties to match the presentation's aesthetic
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Lato', 'Arial']
plt.rcParams['text.color'] = COLOR_PRIMARY_TEXT
plt.rcParams['axes.labelcolor'] = COLOR_PRIMARY_TEXT
plt.rcParams['xtick.color'] = COLOR_PRIMARY_TEXT
plt.rcParams['ytick.color'] = COLOR_PRIMARY_TEXT
plt.rcParams['axes.edgecolor'] = COLOR_GRID
plt.rcParams['axes.titlecolor'] = COLOR_TITLE
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['figure.facecolor'] = COLOR_BG
plt.rcParams['axes.facecolor'] = COLOR_BG


def plot_global_fertility_trend():
    """
    Generates the Area Chart for Global Fertility Rate (Slide 2).
    """
    print("Generating Chart 1: Global Fertility Rate Trend...")
    
    # Data from the presentation
    years = np.array([1950, 1970, 1990, 2010, 2025])
    tfr = np.array([4.8, 4.5, 3.2, 2.5, 2.4])

    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot the line
    ax.plot(years, tfr, color=COLOR_ACCENT_1, marker='o', linewidth=3, markersize=8,
            markeredgecolor=COLOR_ACCENT_1, markerfacecolor=COLOR_BG)

    # Plot the filled area
    ax.fill_between(years, tfr, color=COLOR_ACCENT_1, alpha=0.2)

    # --- Styling ---
    ax.set_title('Global Total Fertility Rate Plummets (1950-2025)', fontsize=18, fontweight='bold', pad=20)
    ax.set_ylabel('Total Fertility Rate (TFR)')
    ax.set_xlabel('Year')
    
    # Set axis limits and ticks
    ax.set_ylim(0, 5.5)
    ax.set_xticks(years)
    ax.set_yticks(np.arange(0, 6, 1))
    
    # Add a horizontal line for the replacement level
    ax.axhline(y=2.1, color=COLOR_ACCENT_3, linestyle='--', linewidth=1.5, label='Replacement Level (2.1)')
    ax.legend(loc='upper right')

    # Add gridlines
    ax.grid(axis='y', color=COLOR_GRID, linestyle='--', linewidth=0.5)
    
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(COLOR_GRID)
    ax.spines['bottom'].set_color(COLOR_GRID)

    plt.tight_layout()
    return fig

def plot_japan_india_comparison():
    """
    Generates the Bar Chart comparing India and Japan (Slide 5).
    """
    print("Generating Chart 2: TFR Comparison (India vs. Japan)...")
    
    # Data from the presentation
    labels = ['Japan (2024)', 'Japan (1980)', 'India (2024)', 'India (1980)']
    values = [1.3, 1.8, 2.0, 4.9]
    colors = [COLOR_ACCENT_1, COLOR_ACCENT_1, COLOR_ACCENT_2, COLOR_ACCENT_2]
    
    # Plot in reverse order for correct display
    labels.reverse()
    values.reverse()
    colors.reverse()

    fig, ax = plt.subplots(figsize=(10, 6))

    # Create horizontal bars
    bars = ax.barh(labels, values, color=colors, height=0.6)
    
    # --- Styling ---
    ax.set_title('Case Study: TFR at a Tipping Point', fontsize=18, fontweight='bold', pad=20)
    ax.set_xlabel('Total Fertility Rate (TFR)')
    
    # Set X-axis limits
    ax.set_xlim(0, 5.5)
    
    # Add data labels on the bars
    ax.bar_label(bars, fmt='%.1f', padding=5, color=COLOR_TITLE, fontsize=12, fontweight='bold')
    
    # Remove top, right, and left spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color(COLOR_GRID)
    
    # Remove y-ticks, keep labels
    ax.tick_params(axis='y', length=0)
    
    # Add gridlines
    ax.grid(axis='x', color=COLOR_GRID, linestyle='--', linewidth=0.5)

    plt.tight_layout()
    return fig

def plot_budget_burden():
    """
    Generates the Doughnut Chart for Government Spending (Slide 9).
    """
    print("Generating Chart 3: Pension & Healthcare Burden...")
    
    # Data from the presentation
    labels = ['Elderly Healthcare', 'Pensions & Social Security', 'Education', 'Other Spending']
    sizes = [45, 30, 10, 15] # Percentages
    colors = [COLOR_ACCENT_1, COLOR_ACCENT_2, COLOR_ACCENT_3, COLOR_ACCENT_4]
    
    fig, ax = plt.subplots(figsize=(10, 7))

    # Create the pie chart
    wedges, texts, autotexts = ax.pie(
        sizes, 
        colors=colors, 
        autopct='%1.0f%%', 
        startangle=90, 
        pctdistance=0.85,
        explode=(0.02, 0.02, 0.02, 0.02) # Slight explosion for separation
    )
    
    # Style percentage labels
    for autotext in autotexts:
        autotext.set_color(COLOR_BAR_TEXT)
        autotext.set_fontsize(12)
        autotext.set_fontweight('bold')

    # Draw a circle in the center to make it a doughnut chart
    centre_circle = plt.Circle((0, 0), 0.70, fc=COLOR_BG)
    fig.gca().add_artist(centre_circle)
    
    # --- Styling ---
    ax.set_title('The Pension & Healthcare Burden', fontsize=18, fontweight='bold', pad=20)
    
    # Add legend
    ax.legend(wedges, labels,
              title="Budget Categories",
              loc="center left",
              bbox_to_anchor=(0.9, 0, 0.5, 1),
              prop={'size': 11})

    ax.axis('equal')  # Equal aspect ratio ensures a circular pie
    
    plt.tight_layout()
    return fig


def main():
    """
    Main function to run the project and display all charts.
    """
    print("--- Starting Data Visualization Project ---")
    
    # Generate all figures
    fig1 = plot_global_fertility_trend()
    fig2 = plot_japan_india_comparison()
    fig3 = plot_budget_burden()
    
    # Show all plots
    print("--- All charts generated successfully. Displaying now. ---")
    plt.show()

if __name__ == "__main__":
    main()
```
